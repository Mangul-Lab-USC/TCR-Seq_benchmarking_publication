```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
est_reads = ['28000','13500','2600','1300','260']
```

MIXCR

```python
df_MIXCR = pd.DataFrame(columns=['th'])

for reads in est_reads: 
    df_temp_aa = pd.read_csv('../summary_data/subsample/capturing_per_reads/capturing_MIXCR_'+reads+'.csv')
    df_temp_aa = df_temp_aa[df_temp_aa['observed_portion_frequency'].notna()]
    df_temp_aa = df_temp_aa[df_temp_aa.observed_portion_frequency != 0]
    df_temp_aa.rename(columns={'observed_portion_frequency':'observed_portion_frequency_'+reads}, inplace=True)
    
    df_MIXCR = pd.merge(df_MIXCR, df_temp_aa, how='right', on=['th'])
    df_MIXCR['tool'] = 'MIXCR'

df_MIXCR
```

IMREP

TRUST4

```python
df_TRUST4 = pd.DataFrame(columns=['th'])

for reads in est_reads: 
    df_temp_aa = pd.read_csv('../summary_data/subsample/capturing_per_reads/capturing_TRUST4_'+reads+'.csv')
    df_temp_aa = df_temp_aa[df_temp_aa['observed_portion_frequency'].notna()]
    df_temp_aa = df_temp_aa[df_temp_aa.observed_portion_frequency != 0]
    df_temp_aa.rename(columns={'observed_portion_frequency':'observed_portion_frequency_'+reads}, inplace=True)

    df_TRUST4 = pd.merge(df_TRUST4, df_temp_aa, how='right', on=['th'])
    df_TRUST4['tool'] = 'TRUST4'

df_TRUST4
```

```python
df_combined = pd.concat([df_MIXCR,df_IMREP,df_TRUST4], ignore_index=True, sort=False)
df_combined
```

```python
tools = ['MIXCR','IMREP','TRUST4']
colors = sns.color_palette("muted")

for tool,color in zip (tools,colors):
    
    sns.set_style("white") 
    sns.set_context("talk")  
    
    figure = df_combined.loc[df_combined['tool'] == tool]
    
    fig, ax = plt.subplots(figsize=(5, 4))

    sns.lineplot(figure['th'], figure['observed_portion_frequency_260'],label='260')
    sns.lineplot(figure['th'], figure['observed_portion_frequency_1300'],label='1300')
    sns.lineplot(figure['th'], figure['observed_portion_frequency_2600'],label='2600')
    sns.lineplot(figure['th'], figure['observed_portion_frequency_13500'],label='13500')
    sns.lineplot(figure['th'], figure['observed_portion_frequency_28000'],label='28000')
    
    ax.legend(['260','1300','2600','13500','28000'])
  
    plt.legend(bbox_to_anchor=(0.2,1), frameon=False, ncol=3, handletextpad=0.2, columnspacing=0.8, handlelength=1, prop={'size': 15})
    locs, labels = plt.xticks()
    ax.set_ylabel('Portion of repetoire captured by' + tool)
    ax.set_xlabel('Clonotype frequency')
    plt.title('Estimated reads',y=1.3,fontsize=16)
    ax.set_xlim(0,0.001)
    ax.set_ylim(0.98,1)

    # setting label sizes after creation
    ax.xaxis.label.set_size(16)
    ax.yaxis.label.set_size(16)
    ax.tick_params(axis='both', which='major', labelsize=13)
    ax.tick_params(axis='both', which='minor', labelsize=13)
    plt.xticks(rotation=45)
    sns.despine()
    plt.subplots_adjust(wspace=0.5)
    plt.show(block=False)
```
