```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv("../summary_data/150bp/all_tools_TRB.csv")
df
```

```python
skin_samples = df.loc[((df['sample_type'] == 'melanoma'))]
skin_samples
```

```python
major = skin_samples.loc[((skin_samples['frequency_TCR'] >= 0.01))]
major
```

```python
# Count the number of clonotypes that are captured by RNA-Seq in each sample

# TCR 
TCR = major.groupby('Sample')['frequency_TCR'].count().reset_index().rename(columns={'frequency_TCR':'count_TCR'})

# MIXCR
MIXCR = major[major['frequency_MIXCR'] != 0].groupby('Sample')['frequency_MIXCR'].count().reset_index().rename(columns={'frequency_MIXCR':'count_tool'})
MIXCR = pd.merge(TCR, MIXCR, how='outer', on=['Sample']).fillna(0)
MIXCR['tool'] = 'MIXCR'

# IMREP
IMREP = major[major['frequency_IMREP'] != 0].groupby('Sample')['frequency_IMREP'].count().reset_index().rename(columns={'frequency_IMREP':'count_tool'})
IMREP = pd.merge(TCR, IMREP, how='outer', on=['Sample']).fillna(0)
IMREP['tool'] = 'IMREP'

# TRUST4
TRUST4 = major[major['frequency_TRUST4'] != 0].groupby('Sample')['frequency_TRUST4'].count().reset_index().rename(columns={'frequency_TRUST4':'count_tool'})
TRUST4 = pd.merge(TCR, TRUST4, how='outer', on=['Sample']).fillna(0)
TRUST4['tool'] = 'TRUST4'

overlap_count = pd.concat([MIXCR,IMREP,TRUST4],ignore_index=True)
overlap_count['overlap_percentage'] = overlap_count['count_tool']/overlap_count['count_TCR']*100

overlap_count
```

```python
overlap_count['overlap_percentage'].mean()
```

```python
tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools:  
    print(tool)
    df_tool = overlap_count.loc[overlap_count['tool'] == tool]
    mean_overlap_count = df_tool['overlap_percentage'].mean()
    display(mean_overlap_count)
```

```python
ax = sns.barplot(x='tool',y='overlap_percentage',data=overlap_count)

ax.set_xlabel('tool',fontsize=10)
ax.set_ylabel('clonotype detection percentage',fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
