```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv("../summary_data/original/all_tools_TRB.csv")
df
```

```python
dominant_clonotype = df.loc[((df['frequency_TCR'] >= 0.1))]
```

```python
# Calculate the abundance of the dominant clonotypes among all clonotypes
TCR_abundance = dominant_clonotype.groupby(['Sample','tissue','tissue_type','repertoire_type','class']).agg({'frequency_TCR':'sum'}).reset_index().rename(columns={'frequency_TCR':"clonotype_abundance"})
TCR_abundance.loc[:,'clonotype_abundance']=TCR_abundance['clonotype_abundance']*100
TCR_abundance
```

```python
print("Mean")
print (TCR_abundance['clonotype_abundance'].mean())

print("STD")
print (TCR_abundance['clonotype_abundance'].std())

mean_clonotype_abundance = TCR_abundance.groupby("class")["clonotype_abundance"].agg(["mean", "std"]) 
display(mean_clonotype_abundance)
```

```python
# Calculate the percentage of the dominant clonotype among all clonotypes (based on clonotype counts)
TCR = T_poor_tissues.groupby(['Sample','repertoire_type','tissue_type'])['frequency_TCR'].count().reset_index().rename(columns={'frequency_TCR':'count_TCR'})
clonotype_count = df[df['frequency_TCR'] != 0].groupby(['Sample','repertoire_type','tissue_type'])['frequency_TCR'].count().reset_index(name='clonotype_count_TCR')

TCR_percentage = pd.merge(TCR, clonotype_count, on=['Sample','repertoire_type','tissue_type'])
TCR_percentage['clonotype_percentage'] = TCR_percentage['count_TCR']/TCR_percentage['clonotype_count_TCR']*100

TCR_percentage
```

```python
print("Mean")
print (TCR_percentage['clonotype_percentage'].mean())

print("STD")
print (TCR_percentage['clonotype_percentage'].std())

mean_TCR_percentage = TCR_percentage.groupby("tissue_type")["clonotype_percentage"].agg(["mean", "std"]) 
display(mean_TCR_percentage)
```

```python
# Calculate the coverage of the dominant clonotype by each tool 

# MIXCR
MIXCR = T_poor_tissues[T_poor_tissues['frequency_MIXCR'] != 0].groupby(['Sample','repertoire_type','tissue_type'])['frequency_MIXCR'].count().reset_index().rename(columns={'frequency_MIXCR':'count_tool'})
MIXCR = pd.merge(TCR, MIXCR, how='outer', on=['Sample','repertoire_type','tissue_type']).fillna(0)
MIXCR['tool'] = 'MIXCR'

# IMREP
IMREP = T_poor_tissues[T_poor_tissues['frequency_IMREP'] != 0].groupby(['Sample','repertoire_type','tissue_type'])['frequency_IMREP'].count().reset_index().rename(columns={'frequency_IMREP':'count_tool'})
IMREP = pd.merge(TCR, IMREP, how='outer', on=['Sample','repertoire_type','tissue_type']).fillna(0)
IMREP['tool'] = 'IMREP'

# TRUST4
TRUST4 = T_poor_tissues[T_poor_tissues['frequency_TRUST4'] != 0].groupby(['Sample','repertoire_type','tissue_type'])['frequency_TRUST4'].count().reset_index().rename(columns={'frequency_TRUST4':'count_tool'})
TRUST4 = pd.merge(TCR, TRUST4, how='outer', on=['Sample','repertoire_type','tissue_type']).fillna(0)
TRUST4['tool'] = 'TRUST4'

overlap_count = pd.concat([MIXCR,IMREP,TRUST4],ignore_index=True)
overlap_count['overlap_percentage'] = overlap_count['count_tool']/overlap_count['count_TCR']*100

overlap_count
```

```python
mean_overlap_percentage = overlap_count.groupby("tissue_type")["overlap_percentage"].agg(["mean", "std"]) 
display(mean_overlap_percentage)
tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools: 
    print(tool)
    df_tool = overlap_count.loc[overlap_count['tool'] == tool]
    mean_overlap_percentage = df_tool.groupby("tissue_type")["overlap_percentage"].agg(["mean", "std"]) 
    display(mean_overlap_percentage)
```

```python
ax = sns.barplot(x='tool',y='overlap_percentage',hue='tissue_type',data=overlap_count,palette="muted")

ax.set_xlabel('tool',fontsize=10)
ax.set_ylabel('clonotype detection percentage',fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
