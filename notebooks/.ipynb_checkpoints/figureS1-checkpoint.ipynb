```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../summary_data/original/all_tools_TRB_reads.csv")
df.drop(['total_reads_tool', 'tool','TCR_derived_by_RNA_seq_tool'], axis=1, inplace=True)
df = df.drop_duplicates(keep='first')
df
```

```python
# Number of RNA-Seq reads in different tissue types
df.loc[:,'total_reads_RNA_seq(million)'] = df['total_reads_RNA_seq']/1000000

sns.set_style("white") 
sns.set_context("talk")  

ax = plt.subplots(figsize = (5,5))

ax = sns.barplot(x='tissue',y='total_reads_RNA_seq(million)',data=df,palette='muted',
                 order=['PBMC','lymph_node','melanoma','small_intestine','kidney'],alpha=0.7)
sns.stripplot(x="tissue",y="total_reads_RNA_seq(million)",data=df,s=5,color=".3",palette="muted",
              dodge=True,linewidth=1,order=['PBMC','lymph_node','melanoma','small_intestine','kidney'])

ax.set_xlabel('Tissue type',fontsize=15)
ax.set_ylabel('RNA-Seq reads (million)',fontsize=15)
plt.xticks(fontsize=15, rotation=45)
plt.yticks(fontsize=15)

sns.despine()

plt.savefig("../figures/supplementary/figureS1a.pdf",bbox_inches="tight")
plt.savefig("../figures/supplementary/figureS1a.png",bbox_inches="tight")
plt.show()
```

```python
# Number of TCR-Seq reads in different tissue types
df.loc[:,'total_reads_TCR(million)'] = df['total_reads_TCR']/1000000

sns.set_style("white") 
sns.set_context("talk")  

ax = plt.subplots(figsize = (5,5))

ax = sns.barplot(x='tissue',y='total_reads_TCR(million)',data=df,palette='muted',
                 order=['PBMC','lymph_node','melanoma','small_intestine','kidney'],alpha=0.7)
sns.stripplot(x="tissue",y="total_reads_TCR(million)",data=df,s=5,color=".3",palette="muted",
              dodge=True,linewidth=1,order=['PBMC','lymph_node','melanoma','small_intestine','kidney'])

ax.set_xlabel('Tissue type',fontsize=15)
ax.set_ylabel('TCR-Seq reads (million)',fontsize=15)
plt.xticks(fontsize=15, rotation=45)
plt.yticks(fontsize=15)

sns.despine()

plt.savefig("../figures/supplementary/figureS1b.pdf",bbox_inches="tight")
plt.savefig("../figures/supplementary/figureS1b.png",bbox_inches="tight")
plt.show()
```

```python
clonotype_TCR = pd.read_csv("../summary_data/original/all_tools_TRB_diversity.csv")
clonotype_TCR = clonotype_TCR[['Sample','clonotype_count_TCR','repertoire_type','class','tissue']].drop_duplicates(keep='first')
clonotype_TCR
```

```python
# Number of TCR clonotype counts in different tissue types
sns.set_style("white") 
sns.set_context("talk")  

ax = plt.subplots(figsize = (5,5))

ax = sns.barplot(x='tissue',y='clonotype_count_TCR',data=clonotype_TCR,palette="muted",alpha=0.7,
                order=['PBMC','lymph_node','melanoma','small_intestine','kidney'])
sns.stripplot(x="tissue",y="clonotype_count_TCR",data=clonotype_TCR,s=5,color=".3",palette="muted",
              dodge=True,linewidth=1,order=['PBMC','lymph_node','melanoma','small_intestine','kidney'])

ax.set_xlabel('Tissue type',fontsize=15)
ax.set_ylabel('TCR-Seq clonotype counts',fontsize=15)
plt.xticks(fontsize=15, rotation=45)
plt.yticks(fontsize=15)
sns.despine()

plt.savefig("../figures/supplementary/figureS1c.pdf",bbox_inches="tight")
plt.savefig("../figures/supplementary/figureS1c.png",bbox_inches="tight")
plt.show()
```
