```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
IMREP_df = pd.read_csv("../summary_data/original/IMREP/IMREP_TRB_merged_extracted_features.csv")
TRUST4_df = pd.read_csv("../summary_data/original/TRUST4/TRUST4_TRB_merged_extracted_features.csv")
MIXCR_df = pd.read_csv("../summary_data/original/MIXCR/MIXCR_TRB_merged_extracted_features.csv")
TCR_df = pd.read_csv("../summary_data/original/TCR_Seq/TCR_merged_extracted_features.csv")
```

```python
# Rename TCR SAMPLE names based on 1:1 matchup in python dictionary
Sample_name_matchup_RNA_TCR_dict = {
    "ESO1-sorted-T-cells_S13_L007" : "RearrangementDetails_ESO1_sorted_infusion",
    "INY1-sorted-T-cells_S14_L007" : "RearrangementDetails_INY1_sorted_infusion",
    "INY2-sorted-T-cells_S15_L007" : "RearrangementDetails_INY2_sorted_infusion",
    "TR2-PBMC_S12" : "RearrangementDetails_TR-PBMC",
    "SAR-11-14-12RNA_S1" : "SAR_11-14_PBMC",
    "MP-11-28-12RNA_S2" : "MP_11-28_PBMC",
    "CMT-baseline1C_CAGATC" : "Pt204_Baseline_TCR_seq",
    "HM-baseline1C_CGATGT" : "Pt310_baseline_TCRseq",
    "PT0310_S9" : "Pt310_on-tx_TCR_seq",
    "LEK-baseline_CGATGT" : "Pt294_baseline_TCR_seq",
    "LEK-OT110712A_CCGTCC" : "Pt294_on-tx_TCR_seq",
    "JSSBaseline-RNA_GTGAAA" : "Pt_308_baseline_TCR_seq",
    "RAS-baseline_TGACCA" : "Pt_325_baseline_TCR_seq",
    "PT0112-B_S3" : "Pt_112_baseline_TCR_seq",
    "PT0285-B_S5" : "Pt_285_baseline_TCR_seq"
}

for matchup in Sample_name_matchup_RNA_TCR_dict:
    TCR_df.loc[(TCR_df['Sample'] == Sample_name_matchup_RNA_TCR_dict[matchup]), 'Sample' ] = matchup
```

```python
# Select rows with nReads greater than 1
IMREP_df = IMREP_df[IMREP_df['nReads'] > 1]
TRUST4_df = TRUST4_df[TRUST4_df['nReads'] > 1]
MIXCR_df = MIXCR_df[MIXCR_df['nReads'] > 1]
TCR_df = TCR_df[TCR_df['nReads'] > 1]
```

```python
# Rename nReads and frequencies columns according to tool for proper merging
IMREP_df = IMREP_df.rename(columns={"nReads": "nReads_IMREP"})
TRUST4_df = TRUST4_df.rename(columns={"nReads": "nReads_TRUST4"})
MIXCR_df = MIXCR_df.rename(columns={"nReads": "nReads_MIXCR"})
TCR_df = TCR_df.rename(columns={"nReads": "nReads_TCR"})
```

```python
# Merge dataframes based on two key combination: Sample and CDR3. Outer join ensures no data is lost for instances that do not have overlap
merge_IMREP_TRUST4 = pd.merge(IMREP_df, TRUST4_df, how='outer', on=['Sample', 'CDR3']).fillna(0)
merge_IMREP_TRUST4_MIXCR = pd.merge(MIXCR_df, merge_IMREP_TRUST4, how='outer', on=['Sample', 'CDR3']).fillna(0)
merge_complete = pd.merge(TCR_df, merge_IMREP_TRUST4_MIXCR, how='outer', on=['Sample', 'CDR3']).fillna(0)

merge_complete
```

```python
# Select monoclonal repertoires
monoclonal = merge_complete.loc[merge_complete['Sample'].isin(['ESO1-sorted-T-cells_S13_L007','INY1-sorted-T-cells_S14_L007','INY2-sorted-T-cells_S15_L007'])]
monoclonal
```

```python
# Calculate total number of reads in each sample
total_reads = monoclonal[['Sample','nReads_TCR','nReads_MIXCR','nReads_IMREP','nReads_TRUST4']].groupby('Sample').sum().rename(columns={'nReads_TCR':'total_reads_TCR','nReads_MIXCR':'total_reads_MIXCR','nReads_IMREP':'total_reads_IMREP','nReads_TRUST4':'total_reads_TRUST4'})
total_reads
```

```python
# Merge dataframes 
merge = pd.merge(monoclonal, total_reads, how='outer', on=['Sample']).fillna(0)

# Calculate frequency of CDR3 reads with respect to CDR3s that occur more than once
merge['frequency_TCR'] = merge['nReads_TCR'] / (merge['total_reads_TCR'] * 1.0)
merge['frequency_MIXCR'] = merge['nReads_MIXCR'] / (merge['total_reads_MIXCR'] * 1.0)
merge['frequency_IMREP'] = merge['nReads_IMREP'] / (merge['total_reads_IMREP'] * 1.0)
merge['frequency_TRUST4'] = merge['nReads_TRUST4'] / (merge['total_reads_TRUST4'] * 1.0)  
merge.fillna(0, inplace=True)
merge
```

Computationally reduce the number of reads in MIXCR (146), IMREP (180), TRUST4 (607)

```python
samples = ['ESO1-sorted-T-cells_S13_L007','INY1-sorted-T-cells_S14_L007','INY2-sorted-T-cells_S15_L007']
df = pd.DataFrame()
for sample in samples: 
    sample = merge.loc[merge['Sample']==sample]
    sample.loc[:,'new_MIXCR_reads'] = 146*sample['frequency_MIXCR']
    sample.loc[:,'new_MIXCR_reads'] = sample['new_MIXCR_reads'].round().astype(int)
    sample.loc[:,'total_MIXCR_reads'] = sample['new_MIXCR_reads'].sum()
    sample.loc[:,'new_IMREP_reads'] = 180*sample['frequency_IMREP']
    sample.loc[:,'new_IMREP_reads'] = sample['new_IMREP_reads'].round().astype(int)
    sample.loc[:,'total_IMREP_reads'] = sample['new_IMREP_reads'].sum()
    sample.loc[:,'new_TRUST4_reads'] = 607*sample['frequency_TRUST4']
    sample.loc[:,'new_TRUST4_reads'] = sample['new_TRUST4_reads'].round().astype(int)
    sample.loc[:,'total_TRUST4_reads'] = sample['new_TRUST4_reads'].sum()
    sample = sample[['Sample','CDR3','new_MIXCR_reads','frequency_MIXCR','total_MIXCR_reads','new_IMREP_reads','frequency_IMREP','total_IMREP_reads','new_TRUST4_reads','frequency_TRUST4','total_TRUST4_reads']]
    df =pd.concat([df, sample])

TCR_merge = merge[['Sample','CDR3','nReads_TCR','total_reads_TCR','frequency_TCR']]
df = pd.merge(df, TCR_merge, on=["Sample",'CDR3'],how='inner')

df
```

```python
g = sns.set_style("white")
g = sns.set_context("paper")
g = sns.lmplot(data=df,x="frequency_TCR",y="frequency_MIXCR",ci=None,scatter_kws={"s": 5, 'alpha':0.5,'color': '#4878d0'},line_kws={'color': '#4878d0'})
g = sns.lmplot(data=df,x="frequency_TCR",y="frequency_IMREP",ci=None,scatter_kws={"s": 5, 'alpha':0.5,'color': '#ee854a'},line_kws={'color': '#ee854a'})
g = sns.lmplot(data=df,x="frequency_TCR",y="frequency_TRUST4",ci=None,scatter_kws={"s": 5, 'alpha':0.5,'color': '#6acc64'},line_kws={'color': '#6acc64'})
#g.set_axis_labels("TCR-Seq-based TCRB clonotype frequency","RNA-Seq-based TCRB clonotype frequency")
#g.fig.suptitle(tool)
g.set(ylim=(0, 1))
g.set(xlim=(0, 1))
```

```python
# MIXCR
diversity_MIXCR = df[['Sample','CDR3','frequency_MIXCR']]
diversity_MIXCR = diversity_MIXCR[diversity_MIXCR.frequency_MIXCR != 0]
diversity_MIXCR['shannon_index'] = -(diversity_MIXCR['frequency_MIXCR']*np.log(diversity_MIXCR['frequency_MIXCR']))
shannon_MIXCR = diversity_MIXCR.groupby(['Sample']).agg({'shannon_index':'sum'}).reset_index().rename(columns={'':"shannon_index"})
shannon_MIXCR['shannon_index_tool_modified'] = np.exp(shannon_MIXCR['shannon_index'])
shannon_MIXCR = shannon_MIXCR[['Sample', 'shannon_index_tool_modified']]
shannon_MIXCR['tool'] = 'MIXCR'

# IMREP
diversity_IMREP = df[['Sample','CDR3','frequency_IMREP']]
diversity_IMREP = diversity_IMREP[diversity_IMREP.frequency_IMREP != 0]
diversity_IMREP['shannon_index'] = -(diversity_IMREP['frequency_IMREP']*np.log(diversity_IMREP['frequency_IMREP']))
shannon_IMREP = diversity_IMREP.groupby(['Sample']).agg({'shannon_index':'sum'}).reset_index().rename(columns={'':"shannon_index"})
shannon_IMREP['shannon_index_tool_modified'] = np.exp(shannon_IMREP['shannon_index'])
shannon_IMREP = shannon_IMREP[['Sample', 'shannon_index_tool_modified']]
shannon_IMREP['tool'] = 'IMREP'

# TRUST4
diversity_TRUST4 = df[['Sample','CDR3','frequency_TRUST4']]
diversity_TRUST4 = diversity_TRUST4[diversity_TRUST4.frequency_TRUST4 != 0]
diversity_TRUST4['shannon_index'] = -(diversity_TRUST4['frequency_TRUST4']*np.log(diversity_TRUST4['frequency_TRUST4']))
shannon_TRUST4 = diversity_TRUST4.groupby(['Sample']).agg({'shannon_index':'sum'}).reset_index().rename(columns={'':"shannon_index"})
shannon_TRUST4['shannon_index_tool_modified'] = np.exp(shannon_TRUST4['shannon_index'])
shannon_TRUST4 = shannon_TRUST4[['Sample', 'shannon_index_tool_modified']]
shannon_TRUST4['tool'] = 'TRUST4'

diversity = pd.concat([shannon_MIXCR,shannon_IMREP,shannon_TRUST4])
diversity
```

```python
df_original = pd.read_csv("../summary_data/original/all_tools_TRB.csv")
df_original = df_original.loc[df_original['Sample'].isin(['ESO1-sorted-T-cells_S13_L007','INY1-sorted-T-cells_S14_L007','INY2-sorted-T-cells_S15_L007'])]
diversity_original = df_original[['Sample','CDR3','frequency_TCR']]
diversity_original = diversity_original[diversity_original.frequency_TCR != 0]
diversity_original['shannon_index'] = -(diversity_original['frequency_TCR']*np.log(diversity_original['frequency_TCR']))
shannon_original = diversity_original.groupby(['Sample']).agg({'shannon_index':'sum'}).reset_index().rename(columns={'':"shannon_index"})
shannon_original['shannon_index_tool_TCR'] = np.exp(shannon_original['shannon_index'])
shannon_original = shannon_original[['Sample', 'shannon_index_tool_TCR']]
shannon_original
```

Merge two dataframes and compare the normalized shannon index

```python
df_compare = pd.merge(diversity, shannon_original, on=["Sample"])
df_compare['absolute_error'] = np.abs(df_compare['shannon_index_tool_modified'] - df_compare['shannon_index_tool_TCR'])
df_compare
```

```python
sns.set_style("white") 
sns.set_context("paper")  

ax = plt.subplots(figsize = (10,10))
ax = sns.boxplot(x='tool',y='absolute_error',data=df_compare,palette="muted")
sns.stripplot(x="tool",y="absolute_error",data=df_compare,s = 5,color=".3",palette="muted",dodge=True,linewidth=1)
ax.set_xlabel('tool',fontsize=15)
ax.set_ylabel('absolute error of TCRB diversity',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(rotation=45)
handles, labels = ax.get_legend_handles_labels()
l = plt.legend(handles[0:3], labels[0:3], bbox_to_anchor=(1, 1))
plt.setp(ax.get_legend().get_texts(), fontsize=15) 
plt.setp(ax.get_legend().get_title(), fontsize=15)
sns.despine()
plt.show()
```
