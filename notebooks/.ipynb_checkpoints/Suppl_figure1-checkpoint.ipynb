```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv("../summary_data/150bp/all_tools_TRB_reads.csv")
df
```

```python
# Total reads of TCR derived by RNA-Seq tools based on clonal type 

ax = plt.subplots(figsize = (5,5))

ax = sns.boxenplot(x='clonal_type',y='total_reads_tool', hue='tool', data=df)

ax.set_xlabel('clonal type',fontsize=10)
ax.set_ylabel('TCR derived by RNA-Seq tools',fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

```python
# Reads of TCR derived by RNA-Seq tools per 1 million RNA-Seq reads based on clonal type

ax = plt.subplots(figsize = (5,5))

ax = sns.boxenplot(x='clonal_type',y='TCR_derived_by_RNA_seq_tool', hue='tool', data=df)

ax.set_xlabel('clonal type',fontsize=10)
ax.set_ylabel('TCR derived by RNA-Seq tools per 1 million RNA-Seq reads',fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

```python
# Total reads of TCR derived by RNA-Seq tools based on tissue type

ax = plt.subplots(figsize = (5,5))

ax = sns.boxenplot(x='sample_type',y='total_reads_tool', hue='tool', data=df)

ax.set_xlabel('sample type',fontsize=10)
ax.set_ylabel('TCR derived by RNA-Seq tools',fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

```python
# Reads of TCR derived by RNA-Seq tools per 1 million RNA-Seq reads based on sample type

ax = plt.subplots(figsize = (5,5))

ax = sns.boxenplot(x='sample_type',y='TCR_derived_by_RNA_seq_tool', hue='tool', data=df)

ax.set_xlabel('clonal type',fontsize=10)
ax.set_ylabel('TCR derived by RNA-Seq tools per 1 million RNA-Seq reads',fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```
