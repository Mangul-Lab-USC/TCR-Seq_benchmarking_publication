```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
repertoire_types = ['T_cell_rich_monoclonal','T_cell_rich_polyclonal','T_cell_poor_polyclonal']
```

MIXCR 

```python
columns = ['th', 'observed_portion_frequency_aa', 'observed_portion_frequency_nt', 'class']
df_MIXCR = pd.DataFrame(columns=columns)

for repertoire in repertoire_types: 
    df_temp_aa = pd.read_csv('../summary_data/original/capturing_aa_ability/capturing_aa_ability_MIXCR_'+repertoire+'.csv')
    df_temp_aa = df_temp_aa[df_temp_aa['observed_portion_frequency'].notna()]
    df_temp_aa = df_temp_aa[df_temp_aa.observed_portion_frequency != 0]
    df_temp_aa.rename(columns={'observed_portion_frequency':'observed_portion_frequency_aa'}, inplace=True)
    
    df_temp_nt = pd.read_csv('../summary_data/original/capturing_nt_ability/capturing_nt_ability_MIXCR_'+repertoire+'.csv')
    df_temp_nt = df_temp_nt[df_temp_nt['observed_portion_frequency'].notna()]
    df_temp_nt = df_temp_nt[df_temp_nt.observed_portion_frequency != 0]
    df_temp_nt.rename(columns={'observed_portion_frequency':'observed_portion_frequency_nt'}, inplace=True)
    
    df_temp_MIXCR = pd.merge(df_temp_aa, df_temp_nt, how='outer', on=['th'])
    df_temp_MIXCR['class'] = repertoire

    df_MIXCR = pd.concat([df_MIXCR, df_temp_MIXCR], ignore_index=True, sort=False)

df_MIXCR
```

```python
for repertoire in repertoire_types:
    
    sns.set_style("white") 
    sns.set_context("paper")  
    
    MIXCR_figure = df_MIXCR.loc[df_MIXCR['class'] == repertoire]
    
    fig, ax = plt.subplots(figsize=(5, 4))
    
    ax.stackplot(MIXCR_figure['th'], MIXCR_figure['observed_portion_frequency_aa'],labels=['aa'],color='#ee854a',alpha=0.5)
    ax.stackplot(MIXCR_figure['th'], MIXCR_figure['observed_portion_frequency_nt'],labels=['nt'],color='#4878d0',alpha=0.5)
    
    locs, labels = plt.xticks()
    ax.set_ylabel('Portion of repetoire captured by MiXCR')
    ax.set_xlabel('Clonotype frequency')
    plt.title(repertoire,y=1.2,fontsize=16)
    plt.legend(bbox_to_anchor=(0.69,1.2), frameon=False, ncol=2, handletextpad=0.2, columnspacing=0.8, handlelength=1, prop={'size': 15})
    #ax.set_xlim(0,0.0005)

    # setting label sizes after creation
    ax.xaxis.label.set_size(16)
    ax.yaxis.label.set_size(16)
    ax.tick_params(axis='both', which='major', labelsize=13)
    ax.tick_params(axis='both', which='minor', labelsize=13)
    plt.xticks(rotation=45)
    sns.despine()
    plt.subplots_adjust(wspace=0.5)
    plt.show(block=False)
```
