```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv("../summary_data/original/all_tools_capturing_ability.csv")
df
```

Plot the portion of the TCRB repertoire that each tool can capture based on the TCR clonotype frequency in each sample (cumulative)

```python
pal = sns.color_palette("muted")
print(pal.as_hex())
```

```python
colors = ['#4878d0','#ee854a','#6acc64']
tools = ['MIXCR','IMREP','TRUST4']      
samples = ['ESO1-sorted-T-cells_S13_L007','INY1-sorted-T-cells_S14_L007','INY2-sorted-T-cells_S15_L007','MP-11-28-12RNA_S2','SAR-11-14-12RNA_S1','TR2-PBMC_S12',
           'CMT-baseline1C_CAGATC', 'HM-baseline1C_CGATGT','JSSBaseline-RNA_GTGAAA','LEK-OT110712A_CCGTCC','LEK-baseline_CGATGT','PT0112-B_S3','PT0285-B_S5','PT0310_S9'
            'RAS-baseline_TGACCA','TCGA-CZ-4862','TCGA-CZ-5463','TCGA-CZ-5985','SRR5233639','SRR5233637']

for sample in samples:
    capture_ability = df.loc[(df['Sample'] == sample)]
    
    for tool, color in zip(tools,colors):
        tool_capture_ability = capture_ability.loc[(capture_ability['tool'] == tool)]
        fig, ax = plt.subplots(figsize=(5, 4))
        ax.stackplot(tool_capture_ability['frequency_TCR'], tool_capture_ability['portion_TCR_cum'],color=color)
        ax.set_xlabel("TCRB clonotype frequency")
        ax.set_ylabel("Portion of TCRB repertoire captured by tool")
        ax.set(title = tool + ' + ' + sample)
        #ax.set_xlim(0,0.001)
        ax.set_ylim(0,100)
        plt.show()
```
