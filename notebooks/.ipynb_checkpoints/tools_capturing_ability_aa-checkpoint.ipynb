```python
import pandas as pd
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv("../summary_data/original/all_tools_TRB.csv")
df
```

Calculate the ability of capturing TCR-Seq comfirmed TCRB clonotypes of each RNA-Seq based methods

```python
samples = ['CMT-baseline1C_CAGATC' , 'ESO1-sorted-T-cells_S13_L007',
           'HM-baseline1C_CGATGT' , 'INY1-sorted-T-cells_S14_L007',
           'INY2-sorted-T-cells_S15_L007' , 'JSSBaseline-RNA_GTGAAA',
           'LEK-OT110712A_CCGTCC' , 'LEK-baseline_CGATGT' , 'MP-11-28-12RNA_S2',
           'PT0112-B_S3' , 'PT0285-B_S5' , 'PT0310_S9' , 'RAS-baseline_TGACCA',
           'SAR-11-14-12RNA_S1' , 'TR2-PBMC_S12', 'SRR5233637', 'SRR5233639',
           'TCGA-CZ-4862','TCGA-CZ-5463','TCGA-CZ-5985']
```

```python
## MIXCR
MIXCR = df[['Sample','tissue_type','repertoire_type','CDR3','frequency_TCR','frequency_MIXCR']]

for sample in samples:
    MIXCR.loc[(MIXCR['frequency_MIXCR'] == 0), 'frequency_tool_TCR'] = 0
    MIXCR.loc[~(MIXCR['frequency_MIXCR'] == 0), 'frequency_tool_TCR'] = MIXCR['frequency_TCR']

# calculate the ability of capturing (non-cumulative)
MIXCR.loc[:,'frequency_TCR_copy'] = MIXCR['frequency_TCR']
MIXCR_sum = MIXCR.groupby(['Sample','tissue_type','repertoire_type','frequency_TCR'])['frequency_tool_TCR'].sum().reset_index()
MIXCR_TCR_sum = MIXCR.groupby(['Sample','tissue_type','repertoire_type','frequency_TCR'])['frequency_TCR_copy'].sum().reset_index(name='frequency_TCR_sum')
MIXCR_TCR = pd.merge(MIXCR_sum, MIXCR_TCR_sum, how='outer', on=['Sample','tissue_type','repertoire_type','frequency_TCR'])
MIXCR_TCR.loc[:,'portion_TCR'] = (MIXCR_TCR['frequency_tool_TCR']*100)/MIXCR_TCR['frequency_TCR_sum']
MIXCR_TCR.fillna(0,inplace=True)

# calculate the ability of capturing (cumulative)
MIXCR_TCR.loc[:,'frequency_tool_TCR_cumulative'] = MIXCR_TCR.groupby('Sample')['frequency_tool_TCR'].cumsum()
MIXCR_TCR = MIXCR_TCR.groupby('Sample').apply(lambda x: x.sort_values('frequency_TCR', ascending = False))
MIXCR_TCR = MIXCR_TCR.droplevel('Sample')  
MIXCR_TCR.loc[:,'portion_TCR_sum'] = MIXCR_TCR.groupby('Sample')['portion_TCR'].cumsum()
MIXCR_TCR['full_coverage'] = (MIXCR_TCR.groupby('Sample').cumcount() + 1)*100
MIXCR_TCR['portion_TCR_cum'] = (MIXCR_TCR['portion_TCR_sum']*100)/MIXCR_TCR['full_coverage']
MIXCR_TCR['tool'] = 'MIXCR'

## IMREP
IMREP = df[['Sample','tissue_type','repertoire_type','CDR3','frequency_TCR','frequency_IMREP']]

for sample in samples:
    IMREP.loc[(IMREP['frequency_IMREP'] == 0), 'frequency_tool_TCR'] = 0
    IMREP.loc[~(IMREP['frequency_IMREP'] == 0), 'frequency_tool_TCR'] = IMREP['frequency_TCR']
    
# calculate the ability of capturing (non-cumulative)
IMREP.loc[:,'frequency_TCR_copy'] = IMREP['frequency_TCR']
IMREP_sum = IMREP.groupby(['Sample','tissue_type','repertoire_type','frequency_TCR'])['frequency_tool_TCR'].sum().reset_index()
IMREP_TCR_sum = IMREP.groupby(['Sample','tissue_type','repertoire_type','frequency_TCR'])['frequency_TCR_copy'].sum().reset_index(name='frequency_TCR_sum')
IMREP_TCR = pd.merge(IMREP_sum, IMREP_TCR_sum, how='outer', on=['Sample','tissue_type','repertoire_type','frequency_TCR'])
IMREP_TCR['portion_TCR'] = (IMREP_TCR['frequency_tool_TCR']*100)/IMREP_TCR['frequency_TCR_sum']
IMREP_TCR.fillna(0,inplace=True)

# calculate the ability of capturing (cumulative)
IMREP_TCR.loc[:,'frequency_tool_TCR_cumulative'] = IMREP_TCR.groupby('Sample')['frequency_tool_TCR'].cumsum()
IMREP_TCR = IMREP_TCR.groupby('Sample').apply(lambda x: x.sort_values('frequency_TCR', ascending = False))
IMREP_TCR = IMREP_TCR.droplevel('Sample')  
IMREP_TCR.loc[:,'portion_TCR_sum'] = IMREP_TCR.groupby('Sample')['portion_TCR'].cumsum()
IMREP_TCR['full_coverage'] = (IMREP_TCR.groupby('Sample').cumcount() + 1)*100
IMREP_TCR['portion_TCR_cum'] = (IMREP_TCR['portion_TCR_sum']*100)/IMREP_TCR['full_coverage']
IMREP_TCR['tool'] = 'IMREP'

## TRUST4
TRUST4 = df[['Sample','tissue_type','repertoire_type','CDR3','frequency_TCR','frequency_TRUST4']]

for sample in samples:
    TRUST4.loc[(TRUST4['frequency_TRUST4'] == 0), 'frequency_tool_TCR'] = 0
    TRUST4.loc[~(TRUST4['frequency_TRUST4'] == 0), 'frequency_tool_TCR'] = TRUST4['frequency_TCR']
    
# calculate the ability of capturing (non-cumulative)
TRUST4.loc[:,'frequency_TCR_copy'] = MIXCR['frequency_TCR']
TRUST4_sum = TRUST4.groupby(['Sample','tissue_type','repertoire_type','frequency_TCR'])['frequency_tool_TCR'].sum().reset_index()
TRUST4_TCR_sum = TRUST4.groupby(['Sample','tissue_type','repertoire_type','frequency_TCR'])['frequency_TCR_copy'].sum().reset_index(name='frequency_TCR_sum')
TRUST4_TCR = pd.merge(TRUST4_sum, TRUST4_TCR_sum, how='outer', on=['Sample','tissue_type','repertoire_type','frequency_TCR'])
TRUST4_TCR['portion_TCR'] = (TRUST4_TCR['frequency_tool_TCR']*100)/TRUST4_TCR['frequency_TCR_sum']
TRUST4_TCR.fillna(0,inplace=True)

# calculate the ability of capturing (cumulative)
TRUST4_TCR.loc[:,'frequency_tool_TCR_cumulative'] = TRUST4_TCR.groupby('Sample')['frequency_tool_TCR'].cumsum()
TRUST4_TCR = TRUST4_TCR.groupby('Sample').apply(lambda x: x.sort_values('frequency_TCR', ascending = False))
TRUST4_TCR = TRUST4_TCR.droplevel('Sample')  
TRUST4_TCR.loc[:,'portion_TCR_sum'] = TRUST4_TCR.groupby('Sample')['portion_TCR'].cumsum()
TRUST4_TCR['full_coverage'] = (TRUST4_TCR.groupby('Sample').cumcount() + 1)*100
TRUST4_TCR['portion_TCR_cum'] = (TRUST4_TCR['portion_TCR_sum']*100)/TRUST4_TCR['full_coverage']
TRUST4_TCR['tool'] = 'TRUST4'

capture_ability = pd.concat([IMREP_TCR,MIXCR_TCR,TRUST4_TCR])
capture_ability.fillna(0,inplace=True)
capture_ability
```

```python
ccc = capture_ability.loc[capture_ability['Sample'] == 'PT0285-B_S5']
ddd = ccc.loc[ccc['tool'] == 'TRUST4']
ddd
```
