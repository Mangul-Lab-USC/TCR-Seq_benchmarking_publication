```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np
```

Compare TCRB and TCRA number of TCR derived reads per one million RNA-Seq reads 

```python
# load dataframe
df_TCRA_reads = pd.read_csv("../summary_data/150bp/all_tools_TRA_reads.csv")
df_TCRA_reads.columns = ['Sample','TRA_total_reads_tool','total_reads_RNA_seq','tool','TRA_derived_by_RNA_seq_tool']                      

df_TCRB_reads = pd.read_csv("../summary_data/150bp/all_tools_TRB_reads.csv")
df_TCRB_reads = df_TCRB_reads[['Sample','sample_type','clonal_type','total_reads_tool','total_reads_RNA_seq','tool','TCR_derived_by_RNA_seq_tool']]
df_TCRB_reads.columns = ['Sample', 'sample_type', 'clonal_type', 'TRB_total_reads_tool','total_reads_RNA_seq','tool','TRB_derived_by_RNA_seq_tool']   
```

```python
# prepare the reads dataframe 
df_reads = pd.merge(df_TCRA_reads, df_TCRB_reads, how='outer', on=['Sample','tool', 'total_reads_RNA_seq']).fillna(0)
df_reads = df_reads[['Sample','sample_type','clonal_type','tool','TRA_derived_by_RNA_seq_tool','TRB_derived_by_RNA_seq_tool']]
df_reads.rename(columns={'TRA_derived_by_RNA_seq_tool':'TRA','TRB_derived_by_RNA_seq_tool':'TRB'}, inplace=True)
```

```python
# reshape the dataframe for plotting and calculation
df_reads_melt = pd.melt(df_reads, id_vars=['Sample','sample_type','clonal_type','tool'], var_name="chain", value_name="reads_derived_by_RNA_seq_tool")
df_reads_melt
```

```python
# monoclonal
monoclonal = df_reads_melt.loc[df_reads_melt['clonal_type'] == 'monoclonal']
mean_monoclonal_reads = monoclonal.groupby("chain")["reads_derived_by_RNA_seq_tool"].agg(["mean", "std"])
display(mean_monoclonal_reads)
tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools:  
    print(tool)
    df_tool = monoclonal.loc[monoclonal['tool'] == tool]
    mean_monoclonal_reads = df_tool.groupby("chain")["reads_derived_by_RNA_seq_tool"].agg(["mean", "std"]) 
    display(mean_monoclonal_reads)
```

```python
df_group1 = monoclonal[monoclonal['chain'] == 'TRA']
df_group2 = monoclonal[monoclonal['chain'] == 'TRB']
display(stats.ttest_ind(df_group1['reads_derived_by_RNA_seq_tool'], df_group2['reads_derived_by_RNA_seq_tool']))
tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools:  
    print(tool)
    df_tool = monoclonal.loc[monoclonal['tool'] == tool]
    df_group1 = df_tool[df_tool['chain'] == 'TRA']
    df_group2 = df_tool[df_tool['chain'] == 'TRB']
    display(stats.ttest_ind(df_group1['reads_derived_by_RNA_seq_tool'], df_group2['reads_derived_by_RNA_seq_tool']))
```

```python
# polyclonal
polyclonal = df_reads_melt.loc[df_reads_melt['clonal_type'] == 'polyclonal']
mean_polyclonal_reads = polyclonal.groupby("chain")["reads_derived_by_RNA_seq_tool"].agg(["mean", "std"])
display(mean_polyclonal_reads)
tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools:  
    print(tool)
    df_tool = polyclonal.loc[polyclonal['tool'] == tool]
    mean_polyclonal_reads = df_tool.groupby("chain")["reads_derived_by_RNA_seq_tool"].agg(["mean", "std"]) 
    display(mean_polyclonal_reads)
```

```python
df_group1 = polyclonal[polyclonal['chain'] == 'TRA']
df_group2 = polyclonal[polyclonal['chain'] == 'TRB']
display(stats.ttest_ind(df_group1['reads_derived_by_RNA_seq_tool'], df_group2['reads_derived_by_RNA_seq_tool']))
tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools:  
    print(tool)
    df_tool = polyclonal.loc[polyclonal['tool'] == tool]
    df_group1 = df_tool[df_tool['chain'] == 'TRA']
    df_group2 = df_tool[df_tool['chain'] == 'TRB']
    display(stats.ttest_ind(df_group1['reads_derived_by_RNA_seq_tool'], df_group2['reads_derived_by_RNA_seq_tool']))
```

```python
tools = ['MIXCR','IMREP','TRUST4']

for tool in tools:
    figure_reads = df_reads_melt.loc[(df_reads_melt['tool'] == tool)]
    ax = sns.barplot(x='clonal_type', y='reads_derived_by_RNA_seq_tool', hue='chain', data=figure_reads)
    ax.set(title = tool)
    ax.set_xlabel('clonal type',fontsize=10)
    ax.set_ylabel('reads derived from RNA-Seq (per one million RNA-Seq reads)',fontsize=10)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)

    plt.show()
```

Compare TCRB and TCRA clonotype

```python
# load dataframe
df_TCRA_diversity = pd.read_csv("../summary_data/150bp/all_tools_TRA_diversity.csv")
df_TCRA_diversity.columns = ['Sample', 'TRA_shannon_index','TRA_clonotype_count','TRA_normalized_shannon_index','tool','sample_type']

df_TCRB_diversity = pd.read_csv("../summary_data/150bp/all_tools_TRB_diversity.csv")
df_TCRB_diversity = df_TCRB_diversity[['Sample','shannon_index_tool','clonotype_count_tool','normalized_shannon_index_tool','tool','sample_type','clonal_type']]
df_TCRB_diversity.columns = ['Sample', 'TRB_shannon_index','TRB_clonotype_count','TRB_normalized_shannon_index','tool','sample_type','clonal_type']

# merge dataframe
df_diversity = pd.merge(df_TCRA_diversity, df_TCRB_diversity, how='outer', on=['Sample','tool','sample_type']).fillna(0)
df_diversity["clonal_type"].replace({0: "polyclonal"}, inplace=True)
```

```python
# prepare the clonotype dataframe

df_clonotype = df_diversity[['Sample','sample_type','clonal_type','tool','TRA_clonotype_count','TRB_clonotype_count']]
df_clonotype.rename(columns={'TRA_clonotype_count':'TRA','TRB_clonotype_count':'TRB'}, inplace=True)
df_clonotype
```

```python
# reshape the dataframe for plotting and calculation
df_clonotype_melt = pd.melt(df_clonotype, id_vars=['Sample','sample_type','clonal_type','tool'], var_name="chain", value_name="clonotype_counts")
df_clonotype_melt
```

```python
# monoclonal
monoclonal = df_clonotype_melt.loc[df_clonotype_melt['clonal_type'] == 'monoclonal']
mean_monoclonal_clonotype = monoclonal.groupby("chain")["clonotype_counts"].agg(["mean", "std"])
display(mean_monoclonal_clonotype)
tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools:  
    print(tool)
    df_tool = monoclonal.loc[monoclonal['tool'] == tool]
    mean_monoclonal_clonotype = df_tool.groupby("chain")["clonotype_counts"].agg(["mean", "std"]) 
    display(mean_monoclonal_clonotype)
```

```python
df_group1 = monoclonal[monoclonal['chain'] == 'TRA']
df_group2 = monoclonal[monoclonal['chain'] == 'TRB']
display(stats.ttest_ind(df_group1['clonotype_counts'], df_group2['clonotype_counts']))
tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools:  
    print(tool)
    df_tool = monoclonal.loc[monoclonal['tool'] == tool]
    df_group1 = df_tool[df_tool['chain'] == 'TRA']
    df_group2 = df_tool[df_tool['chain'] == 'TRB']
    display(stats.ttest_ind(df_group1['clonotype_counts'], df_group2['clonotype_counts']))
```

```python
# polyclonal
polyclonal = df_clonotype_melt.loc[df_clonotype_melt['clonal_type'] == 'polyclonal']
mean_polyclonal_clonotype = polyclonal.groupby("chain")["clonotype_counts"].agg(["mean", "std"])
display(mean_polyclonal_clonotype)
tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools:  
    print(tool)
    df_tool = polyclonal.loc[polyclonal['tool'] == tool]
    mean_polyclonal_clonotype = df_tool.groupby("chain")["clonotype_counts"].agg(["mean", "std"]) 
    display(mean_polyclonal_clonotype)
```

```python
df_group1 = polyclonal[polyclonal['chain'] == 'TRA']
df_group2 = polyclonal[polyclonal['chain'] == 'TRB']
display(stats.ttest_ind(df_group1['clonotype_counts'], df_group2['clonotype_counts']))
tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools:  
    print(tool)
    df_tool = polyclonal.loc[polyclonal['tool'] == tool]
    df_group1 = df_tool[df_tool['chain'] == 'TRA']
    df_group2 = df_tool[df_tool['chain'] == 'TRB']
    display(stats.ttest_ind(df_group1['clonotype_counts'], df_group2['clonotype_counts']))
```

```python
tools = ['MIXCR','IMREP','TRUST4']

for tool in tools:
    figure_clonotype = df_clonotype_melt.loc[(df_clonotype_melt['tool'] == tool)]
    ax = sns.barplot(x='clonal_type', y='clonotype_counts', hue='chain', data=figure_clonotype)
    ax.set(title = tool)
    ax.set_xlabel('clonal type',fontsize=10)
    ax.set_ylabel('clonotype counts',fontsize=10)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)

    plt.show()
```

Compare TCRB and TCRA SDI

```python
g = sns.FacetGrid(df_diversity, col='clonal_type', row='tool', height=4, margin_titles=True)
g.map(sns.scatterplot, 'TRB_normalized_shannon_index', 'TRA_normalized_shannon_index')
g.set_axis_labels("TRB normalized shannon index", "TRA normalized shannon index")
g.add_legend()
```

```python
tools = ['MIXCR','IMREP','TRUST4']

for tool in tools:
    figure_diversity = df_diversity.loc[(df_diversity['tool'] == tool)]
    ax = sns.scatterplot(x='TRB_normalized_shannon_index',y='TRA_normalized_shannon_index', hue='clonal_type', data=figure_diversity)
    ax.set(title = tool)
    ax.plot([0,1],[0,1],linestyle='dashed')
    ax.set_xlabel('TRB normalized SDI',fontsize=10)
    ax.set_ylabel('TRA normalized SDI',fontsize=10)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)

    plt.show()
```

```python
# Calculate the Pearson correlation 

tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools: 
    correlation = df_diversity.loc[(df_diversity['tool'] == tool)]
    print(tool + ' Pearson correlation cofficient:' + str(stats.pearsonr(correlation['TRA_normalized_shannon_index'], correlation['TRB_normalized_shannon_index'])))
```
