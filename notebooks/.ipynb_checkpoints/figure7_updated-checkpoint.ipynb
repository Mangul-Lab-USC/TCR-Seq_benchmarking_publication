```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np
import itertools
```

```python
df = pd.read_csv("../summary_data/original/all_tools_capturing_ability_nt.csv")
df
```

Calculate the portion of TCR repertoires that each tool can capture

```python
capture_sum = df.groupby(['sample','tissue_type','repertoire_type','tool'])['frequency_tool_TCR'].sum().reset_index(name='capture_sum_by_tool')
capture_sum
```

```python
capture_sum.groupby(['repertoire_type'])['capture_sum_by_tool'].mean().reset_index()
```

```python
capture_sum.groupby(['repertoire_type','tool'])['capture_sum_by_tool'].mean().reset_index()
```

```python
repertoire_types = ['monoclonal','polyclonal']
palette = sns.color_palette(["#4878d0","#6acc64"])

for repertoire_type in repertoire_types:
    repertoire_type_capture_ability = capture_sum.loc[(capture_sum['repertoire_type'] == repertoire_type)]
    
    ax = plt.subplots(figsize = (5,5))

    ax = sns.boxplot(x='tool',y='capture_sum_by_tool', data=repertoire_type_capture_ability, palette=palette)
    sns.stripplot(x="tool", y="capture_sum_by_tool",data=repertoire_type_capture_ability, s = 5, color=".3")

    ax.set_xlabel('tool',fontsize=10)
    ax.set_ylabel('percentage of TCR nucleotide sequence captured',fontsize=10)
    ax.set(title = repertoire_type)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.ylim(0, 1)
    plt.show()
```

Plot the portion of the TCRB repertoire that each tool can capture based on the TCR clonotype frequency in monoclonal repertoire (cumulative)

```python
monoclonal_samples = df.loc[((df['repertoire_type'] == 'monoclonal'))]
palette = itertools.cycle(sns.color_palette(["#4878d0","#6acc64"]))

tools = ['MIXCR','TRUST4']              
for tool in tools: 
    monoclonal_capture_ability = monoclonal_samples.loc[(monoclonal_samples['tool'] == tool)]
    
    fig, ax = plt.subplots(figsize=(5, 4))
    ax.stackplot(monoclonal_capture_ability['frequency_TCR'], monoclonal_capture_ability['portion_TCR_cum'],color=next(palette))
    ax.set_xlabel("TCRB nucleotide sequence frequency")
    ax.set_ylabel("Portion of TCRB nucleotide sequence captured by tool")
    ax.set(title = tool + ' + monoclonal')
    ax.set_xlim(0,0.01)
    plt.show()
```

Plot the portion of the TCRB repertoire that each tool can capture based on the TCR clonotype frequency in polyclonal repertoires (cumulative)

```python
polyclonal_samples = df.loc[((df['repertoire_type'] == 'polyclonal'))]
palette = itertools.cycle(sns.color_palette(["#4878d0","#6acc64"]))

tools = ['MIXCR','TRUST4']              
for tool in tools: 
    polyclonal_capture_ability = polyclonal_samples.loc[(polyclonal_samples['tool'] == tool)]
    
    fig, ax = plt.subplots(figsize=(5, 4))
    ax.stackplot(polyclonal_capture_ability['frequency_TCR'], polyclonal_capture_ability['portion_TCR_cum'],color=next(palette))
    ax.set_xlabel("TCRB nucleotide sequence frequency")
    ax.set_ylabel("Portion of TCRB nucleotide sequence captured by tool")
    ax.set(title = tool + ' + polyclonal')
    ax.set_xlim(0,0.01)
    plt.show()
```
