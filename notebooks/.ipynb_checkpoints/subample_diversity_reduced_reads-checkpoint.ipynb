```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../summary_data/subsample/subsample_diversity.csv")
```

```python
mean_diversity_combined = df.groupby(["tool","est_reads"])["absolute_error"].mean().reset_index()
mean_diversity_combined = mean_diversity_combined.sort_values('est_reads')
mean_diversity_combined
```

```python
sns.set_style('white')
sns.set_context('talk')

fig, ax = plt.subplots(figsize=(10, 10))
ax = sns.lineplot(data=mean_diversity_combined, x="est_reads", y="absolute_error", hue="tool",style="tool",
                  hue_order=['MIXCR','IMREP','TRUST4'],markers=True,ci=None,sort=True)
plt.axhline(y=0.5, color='r', linestyle='--')
ax.set(xlim=(250,30000))
ax.set(ylim=(0,1))
ax.set_xlabel('Total number of derived TCR reads')
ax.set_ylabel('Absolute error of Shannon diversity index')
sns.despine()
```

```python
sns.set_style('white')
sns.set_context('talk')

samples = ['ESO1-sorted-T-cells_S13_L007','INY1-sorted-T-cells_S14_L007','INY2-sorted-T-cells_S15_L007']
for sample in samples: 
    fig, ax = plt.subplots(figsize=(10, 10))
    sample_figure = df.loc[df['Sample']==sample]
    ax = sns.lineplot(data=sample_figure, x="new_total_reads_tool", y="absolute_error", hue="tool",style="tool",markers=True,ci=None)
    ax.set(xlim=(250,30000))
    ax.set(ylim=(0,1))
    ax.set_title(sample)
    ax.set_xlabel('Total number of derived TCR-Seq reads')
    ax.set_ylabel('Absolute error of Shannon diversity index')
    sns.despine()
```
