```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv("../summary_data/original/all_tools_TRB.csv")
df = df.loc[~(df['frequency_TCR'] == 0)]
df
```

```python
# Define clonotype groups based on frequency
def clonotype_group (row):
    if row['frequency_TCR'] > 0.1 and row['frequency_TCR'] <= 1:
        return 'major'
    if row['frequency_TCR'] > 0 and row['frequency_TCR'] <= 0.1: 
        return 'minor'

# Apply the clonotype_group function to the dataframe
df['clonotype_group'] = df.apply (lambda row: clonotype_group(row),axis=1)
df
```

```python
# Calculate the abundance of the dominant clonotypes among all clonotypes
TCR_abundance = df.groupby(['Sample','tissue','tissue_type','repertoire_type','class','clonotype_group']).agg({'frequency_TCR':'sum'}).reset_index().rename(columns={'frequency_TCR':"clonotype_abundance"})
TCR_abundance.loc[:,'clonotype_abundance']=TCR_abundance['clonotype_abundance']*100
TCR_abundance
```

```python
mean_clonotype_abundance = TCR_abundance.groupby(["class","clonotype_group"])["clonotype_abundance"].agg(["mean", "std"]) 
display(mean_clonotype_abundance)
```

```python
sns.set_style("white") 
sns.set_context("paper")  

ax = plt.subplots(figsize = (10,10))
ax = sns.violinplot(x='class',y='clonotype_abundance',data=TCR_abundance,palette="muted",hue='clonotype_group',split=True)
sns.stripplot(x="class",y="clonotype_abundance",data=TCR_abundance,s = 5,palette="muted",hue='clonotype_group',dodge=True,linewidth=1)
ax.set_xlabel('sample type',fontsize=15)
ax.set_ylabel('clonotype abundance (%)',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(rotation=45)
handles, labels = ax.get_legend_handles_labels()
l = plt.legend(handles[0:2], labels[0:2], bbox_to_anchor=(1, 1))
plt.setp(ax.get_legend().get_texts(), fontsize=15) 
plt.setp(ax.get_legend().get_title(), fontsize=15)
sns.despine()
plt.show()
```

```python
# Calculate the coverage of the dominant clonotype by each tool 
# TCR 
TCR = df.groupby(['Sample','tissue','tissue_type','repertoire_type','class','clonotype_group'])['frequency_TCR'].count().reset_index().rename(columns={'frequency_TCR':'count_TCR'})

# MIXCR
MIXCR = df[df['frequency_MIXCR'] != 0].groupby(['Sample','tissue','tissue_type','repertoire_type','class','clonotype_group'])['frequency_MIXCR'].count().reset_index().rename(columns={'frequency_MIXCR':'count_tool'})
MIXCR = pd.merge(TCR, MIXCR, how='outer', on=['Sample','tissue','tissue_type','repertoire_type','class','clonotype_group']).fillna(0)
MIXCR['tool'] = 'MIXCR'
MIXCR

# IMREP
IMREP = df[df['frequency_IMREP'] != 0].groupby(['Sample','tissue','tissue_type','repertoire_type','class','clonotype_group'])['frequency_IMREP'].count().reset_index().rename(columns={'frequency_IMREP':'count_tool'})
IMREP = pd.merge(TCR, IMREP, how='outer', on=['Sample','tissue','tissue_type','repertoire_type','class','clonotype_group']).fillna(0)
IMREP['tool'] = 'IMREP'

# TRUST4
TRUST4 = df[df['frequency_TRUST4'] != 0].groupby(['Sample','tissue','tissue_type','repertoire_type','class','clonotype_group'])['frequency_TRUST4'].count().reset_index().rename(columns={'frequency_TRUST4':'count_tool'})
TRUST4 = pd.merge(TCR, TRUST4, how='outer', on=['Sample','tissue','tissue_type','repertoire_type','class','clonotype_group']).fillna(0)
TRUST4['tool'] = 'TRUST4'

overlap_count = pd.concat([MIXCR,IMREP,TRUST4],ignore_index=True)
overlap_count['overlap_percentage'] = overlap_count['count_tool']/overlap_count['count_TCR']*100

overlap_count
```

```python
mean_overlap_percentage = overlap_count.groupby(["class","clonotype_group"])["overlap_percentage"].agg(["mean", "std"]) 
display(mean_overlap_percentage)
tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools: 
    print(tool)
    df_tool = overlap_count.loc[overlap_count['tool'] == tool]
    mean_overlap_percentage = df_tool.groupby(["class","clonotype_group"])["overlap_percentage"].agg(["mean", "std"]) 
    display(mean_overlap_percentage)
```

```python
clonotypes = ["major","minor"]

sns.set_style("white") 
sns.set_context("paper")  

for clonotype in clonotypes: 
    df_figure = overlap_count.loc[overlap_count['clonotype_group'] == clonotype]
    ax = plt.subplots(figsize = (10,10))
    ax = sns.barplot(x='class',y='overlap_percentage',data=df_figure,palette="muted",hue='tool')
    ax.set_xlabel('sample type',fontsize=15)
    ax.set_ylabel('percentage of captured clonotypes (%)',fontsize=15)
    ax.set_title(clonotype + " clonotype", fontdict={'fontsize': 20, 'fontweight': 'medium'})
    plt.xticks(fontsize=15, rotation=45)
    plt.yticks(fontsize=15)
    ax.set(ylim=(0, 100))
    plt.setp(ax.get_legend().get_texts(), fontsize=15) 
    plt.setp(ax.get_legend().get_title(), fontsize=15)
    sns.despine()
    plt.show()
```
