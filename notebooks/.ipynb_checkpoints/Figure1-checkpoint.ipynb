```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np
import itertools
```

```python
df = pd.read_csv("../summary_data/original/all_tools_capturing_ability.csv")
df
```

Calculate the portion of TCR repertoires that each tool can capture

```python
capture_sum = df.groupby(['Sample','tissue_type','repertoire_type','tool'])['frequency_tool_TCR'].sum().reset_index(name='capture_sum_by_tool')
capture_sum
```

```python
capture_sum.groupby(['repertoire_type'])['capture_sum_by_tool'].mean().reset_index()
```

```python
capture_sum.groupby(['repertoire_type','tool'])['capture_sum_by_tool'].mean().reset_index()
```

```python
repertoire_types = ['monoclonal','polyclonal']

for repertoire_type in repertoire_types:
    repertoire_type_capture_ability = capture_sum.loc[(capture_sum['repertoire_type'] == repertoire_type)]
    
    ax = plt.subplots(figsize = (5,5))

    ax = sns.boxplot(x='tool',y='capture_sum_by_tool', data=repertoire_type_capture_ability, palette="muted")
    sns.stripplot(x="tool", y="capture_sum_by_tool",data=repertoire_type_capture_ability, s = 5, color=".3")

    ax.set_xlabel('tool',fontsize=10)
    ax.set_ylabel('percentage of TCR repertoire captured',fontsize=10)
    ax.set(title = repertoire_type)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)

    plt.show()
```

Plot the portion of the TCRB repertoire that each tool can capture based on the TCR clonotype frequency in monoclonal repertoire (cumulative)

```python
monoclonal_samples = df.loc[((df['repertoire_type'] == 'monoclonal'))]
palette = itertools.cycle(sns.color_palette('muted'))

tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools: 
    monoclonal_capture_ability = monoclonal_samples.loc[(monoclonal_samples['tool'] == tool)]
    
    fig, ax = plt.subplots(figsize=(5, 4))
    ax.stackplot(monoclonal_capture_ability['frequency_TCR'], monoclonal_capture_ability['portion_TCR_cum'],color=next(palette))
    ax.set_xlabel("TCRB clonotype frequency")
    ax.set_ylabel("Portion of TCRB repertoire captured by tool")
    ax.set(title = tool + ' + monoclonal')
    ax.set_xlim(0,0.0003)
    plt.show()
```

Find the TCRB clonotype frequency that reaches 100

```python
monoclonal_samples = df.loc[((df['repertoire_type'] == 'monoclonal'))]
monoclonal_sample_names = ['ESO1-sorted-T-cells_S13_L007','INY1-sorted-T-cells_S14_L007','INY2-sorted-T-cells_S15_L007']

for sample in monoclonal_sample_names: 
    monoclonal_samples_min = monoclonal_samples.loc[(monoclonal_samples['Sample'] == sample)]
    display(monoclonal_samples_min.loc[(monoclonal_samples_min.portion_TCR_cum<=99).groupby(monoclonal_samples_min.tool).idxmax])
```

Plot the portion of the TCRB repertoire that each tool can capture based on the TCR clonotype frequency in polyclonal repertoires (cumulative)

```python
polyclonal_samples = df.loc[((df['repertoire_type'] == 'polyclonal'))]
palette = itertools.cycle(sns.color_palette('muted'))

tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools: 
    polyclonal_capture_ability = polyclonal_samples.loc[(polyclonal_samples['tool'] == tool)]
    
    fig, ax = plt.subplots(figsize=(5, 4))
    ax.stackplot(polyclonal_capture_ability['frequency_TCR'], polyclonal_capture_ability['portion_TCR_cum'],color=next(palette))
    ax.set_xlabel("TCRB clonotype frequency")
    ax.set_ylabel("Portion of TCRB repertoire captured by tool")
    ax.set(title = tool + ' + polyclonal')
    ax.set_xlim(0,0.01)
    plt.show()
```

Plot the portion of the TCRB repertoire that each tool can capture based on the TCR clonotype frequency in different tissue types (cumulative)

```python
colors = ['#4878d0','#ee854a','#6acc64']
tools = ['MIXCR','IMREP','TRUST4']   
tissues = ['PBMC','melanoma','kidney','lymph_node','small_intestine']

for tissue in tissues:
    capture_ability = df.loc[(df['tissue_type'] == tissue)]
    
    for tool, color in zip(tools,colors): 
        tool_capture_ability = capture_ability.loc[(capture_ability['tool'] == tool)]
        fig, ax = plt.subplots(figsize=(5, 4))
        ax.stackplot(tool_capture_ability['frequency_TCR'], tool_capture_ability['portion_TCR_cum'],color=color)
        ax.set_xlabel("TCRB clonotype frequency")
        ax.set_ylabel("Portion of TCRB repertoire captured by tool")
        ax.set(title = tool + ' + ' + tissue)
        ax.set_xlim(0,0.01)
        ax.set_ylim(0,100)
        plt.show()
```
