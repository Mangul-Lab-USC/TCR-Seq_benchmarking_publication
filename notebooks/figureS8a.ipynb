```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../summary_data/subsample/subsample_diversity.csv")
df
```

```python
sns.set_style('white')
sns.set_context('talk')

fig, ax = plt.subplots(figsize=(10, 10))
ax = sns.barplot(data=df, x="tool", y="absolute_error", hue="est_reads", palette="muted")
ax.set(ylim=(0,1))

handles, labels = ax.get_legend_handles_labels()
plt.legend(handles[0:5], labels[0:5], bbox_to_anchor=(0.7, 0.7), frameon=False, title='Estimated reads')

ax.set_xlabel('Tools',fontsize=20)
ax.set_ylabel('Absolute error of SDI',fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.tick_params(axis='both', which='minor', labelsize=20)
plt.setp(ax.get_legend().get_texts(), fontsize=20) 
plt.setp(ax.get_legend().get_title(), fontsize=20)
sns.despine()

plt.savefig("../figures/supplementary/figureS8a.pdf",bbox_inches="tight")
plt.savefig("../figures/supplementary/figureS8a.png",bbox_inches="tight")
```

```python
mean_diversity_combined = df.groupby(["tool","est_reads"])["absolute_error"].mean().reset_index()
mean_diversity_combined = mean_diversity_combined.sort_values('est_reads')
mean_diversity_combined
```
