```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib_venn import venn3
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../summary_data/original/all_tools_TRB.csv')
df = df[df.nReads_TCR != 0]
df
```

```python
# MIXCR clonotypes
MIXCR = df[df.nReads_MIXCR != 0]

# IMREP clonotype
IMREP = df[df.nReads_IMREP != 0]

# TRUST4 clonotype
TRUST4 = df[df.nReads_TRUST4 != 0]
```

## T_cell_rich_monoclonal

```python
ax = plt.subplots(figsize = (5,5))
MIXCR_figure = MIXCR.loc[MIXCR['class'] == 'T_cell_rich_monoclonal']
MIXCR_list = MIXCR_figure['CDR3'].tolist()
print('MIXCR: ', len(MIXCR_list))
IMREP_figure = IMREP.loc[IMREP['class'] == 'T_cell_rich_monoclonal']
IMREP_list = IMREP_figure['CDR3'].tolist()
print('IMREP: ', len(IMREP_list))
TRUST4_figure = TRUST4.loc[TRUST4['class'] == 'T_cell_rich_monoclonal']
TRUST4_list = TRUST4_figure['CDR3'].tolist()
print('TRUST4: ', len(TRUST4_list))

venn3([set(MIXCR_list), set(IMREP_list), set(TRUST4_list)], set_labels=('MIXCR', 'IMREP', 'TRUST4'),alpha = 0.6)
plt.title('T cell rich monoclonal samples')

plt.savefig("../figures/main/figure4_1.pdf",bbox_inches="tight")
plt.savefig("../figures/main/figure4_1.png",bbox_inches="tight")
plt.show()
```
