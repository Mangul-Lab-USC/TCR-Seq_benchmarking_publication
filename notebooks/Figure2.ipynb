```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv("../summary_data/150bp/all_tools_TRB_diversity.csv")
df
```

Normalized SDI comparison

```python
# Violin plot

ax = plt.subplots(figsize = (5,5))

ax = sns.violinplot(x='clonal_type',y='normalized_shannon_index_tool', hue='tool', data=df)

ax.set_xlabel('clonal type',fontsize=10)
ax.set_ylabel('Normalized SDI',fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

```python
# Compare normalized SDI 

tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools: 
    figure_normalized_SDI = df.loc[(df['tool'] == tool)]
    sns.scatterplot(x = 'normalized_shannon_index_TCR', y = 'normalized_shannon_index_tool', hue = 'clonal_type', s = 100, data = figure_normalized_SDI)
    ax = sns.lineplot(x=range(0,2),y=range(0,2))
    ax.set(title = tool)
    ax.lines[0].set_linestyle("--")
    ax.set(xlim = (0, 1), ylim = (0, 1))
    ax.grid(False)
    plt.show()
```

```python
# Calculate the Pearson correlation 

tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools: 
    correlation = df.loc[(df['tool'] == tool)]
    print(tool + ' Pearson correlation cofficient:' + str(stats.pearsonr(correlation['normalized_shannon_index_TCR'], correlation['normalized_shannon_index_tool'])))
```

```python
# Calculate the Pearson correlation for polyclonal samples
polyclonal = df.loc[(df['clonal_type'] == 'polyclonal')]

tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools: 
    correlation = polyclonal.loc[(polyclonal['tool'] == tool)]
    print(tool + ' Pearson correlation cofficient:' + str(stats.pearsonr(correlation['normalized_shannon_index_TCR'], correlation['normalized_shannon_index_tool'])))
```

Relative error of clonal diversity estimation

```python
tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools: 
    method = df.loc[(df['tool'] == tool)]
    relative_error = np.mean(np.abs(method['normalized_shannon_index_TCR'] - method['normalized_shannon_index_tool'])/method['normalized_shannon_index_TCR'])
    print(tool + ' relative error:' + str(relative_error))
```

```python
# Calculate the relative error for monoclonal samples
monoclonal = df.loc[(df['clonal_type'] == 'monoclonal')]

tools = ['MIXCR','IMREP','TRUST4']              
for tool in tools: 
    method = monoclonal.loc[(monoclonal['tool'] == tool)]
    relative_error = np.mean(np.abs(method['normalized_shannon_index_TCR'] - method['normalized_shannon_index_tool'])/method['normalized_shannon_index_TCR'])
    print(tool + ' relative error: ' + str(relative_error))
```
