```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../summary_data/original/all_tools_TRB_reads.csv")
df
```

```python
# Number of TCR derived reads by RNA-Seq based methods per 1 million RNA-Seq reads in different tissue types 
sns.set_style("white") 
sns.set_context("talk")  

ax = plt.subplots(figsize = (10,10))

ax = sns.barplot(x='tissue',y='TCR_derived_by_RNA_seq_tool',hue='tool',data=df,palette='muted',
                 order=['PBMC','lymph_node','melanoma','small_intestine','kidney'],alpha=0.7)
sns.stripplot(x="tissue",y="TCR_derived_by_RNA_seq_tool",data=df,s = 5,color=".3",palette="muted",hue='tool',
              dodge=True,linewidth=1,order=['PBMC','lymph_node','melanoma','small_intestine','kidney'])

ax.set_xlabel('Tissue type',fontsize=15)
ax.set_ylabel('TCR derived reads from RNA-Seq per 1 million RNA-Seq reads (log)',fontsize=15)
ax.set_yscale("log")

handles, labels = ax.get_legend_handles_labels()
l = plt.legend(handles[0:3], labels[0:3], bbox_to_anchor=(1, 1), frameon=False)
plt.setp(ax.get_legend().get_texts(), fontsize=15) 
plt.setp(ax.get_legend().get_title(), fontsize=15)
plt.xticks(fontsize=15, rotation=45)
plt.yticks(fontsize=15)

sns.despine()

plt.savefig("../figures/supplementary/figureS2_a.pdf",bbox_inches="tight")
plt.savefig("../figures/supplementary/figureS2_a.png",bbox_inches="tight")
plt.show()
```

```python
# Number of TCR derived reads by RNA-Seq based methods per 1 million RNA-Seq reads in different repertoire types
repertoire_classes = ['T_cell_rich_monoclonal','T_cell_rich_polyclonal','T_cell_poor_monoclonal','T_cell_poor_polyclonal']
sns.set_style("white") 
sns.set_context("talk")  

ax = plt.subplots(figsize = (10,10))

ax = sns.barplot(x='class',y='TCR_derived_by_RNA_seq_tool',hue='tool',data=df,palette='muted',
                 order=repertoire_classes,alpha=0.7)
sns.stripplot(x="class",y="TCR_derived_by_RNA_seq_tool",data=df,s = 5,color=".3",palette="muted",hue='tool',
              dodge=True,linewidth=1,order=repertoire_classes)

ax.set_xlabel('Repertoire type',fontsize=15)
ax.set_ylabel('TCR derived reads from RNA-Seq per 1 million RNA-Seq reads (log)',fontsize=15)
ax.set_yscale("log")

handles, labels = ax.get_legend_handles_labels()
l = plt.legend(handles[0:3], labels[0:3], bbox_to_anchor=(1, 1), frameon=False)
plt.setp(ax.get_legend().get_texts(), fontsize=15) 
plt.setp(ax.get_legend().get_title(), fontsize=15)
plt.xticks(fontsize=15, rotation=45)
plt.yticks(fontsize=15)

sns.despine()

plt.savefig("../figures/supplementary/figureS2_b.pdf",bbox_inches="tight")
plt.savefig("../figures/supplementary/figureS2_b.png",bbox_inches="tight")
plt.show()
```
