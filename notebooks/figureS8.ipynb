```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
samples = ['sample01','sample02','sample03','sample04','sample05','sample06','sample07','sample08',
           'sample09','sample10','sample11','sample12','sample13','sample14','sample15',
            'SRR5233639','SRR5233637','TCGA-CZ-4862','TCGA-CZ-5463','TCGA-CZ-5985']
```

```python
columns = ['th', 'observed_portion_frequency', 'sample']
df_MIXCR = pd.DataFrame(columns=columns)

for sample in samples: 
    df_temp_aa = pd.read_csv('../summary_data/original/capturing_per_sample/capturing_MIXCR_'+sample+'.csv')
    df_temp_aa = df_temp_aa[df_temp_aa['observed_portion_frequency'].notna()]
    df_temp_aa = df_temp_aa[df_temp_aa.observed_portion_frequency != 0]
    
    df_temp_aa['sample'] = sample

    df_MIXCR = pd.concat([df_MIXCR, df_temp_aa], ignore_index=True, sort=False)
    df_MIXCR['tool'] = 'MIXCR'
```

```python
columns = ['th', 'observed_portion_frequency', 'sample']
df_IMREP = pd.DataFrame(columns=columns)

for sample in samples: 
    df_temp_aa = pd.read_csv('../summary_data/original/capturing_per_sample/capturing_IMREP_'+sample+'.csv')
    df_temp_aa = df_temp_aa[df_temp_aa['observed_portion_frequency'].notna()]
    df_temp_aa = df_temp_aa[df_temp_aa.observed_portion_frequency != 0]
    
    df_temp_aa['sample'] = sample

    df_IMREP = pd.concat([df_IMREP, df_temp_aa], ignore_index=True,sort=False)
    df_IMREP['tool'] = 'IMREP'
```

```python
columns = ['th', 'observed_portion_frequency', 'sample']
df_TRUST4 = pd.DataFrame(columns=columns)

for sample in samples: 
    df_temp_aa = pd.read_csv('../summary_data/original/capturing_per_sample/capturing_TRUST4_'+sample+'.csv')
    df_temp_aa = df_temp_aa[df_temp_aa['observed_portion_frequency'].notna()]
    df_temp_aa = df_temp_aa[df_temp_aa.observed_portion_frequency != 0]
   
    df_temp_aa['sample'] = sample

    df_TRUST4 = pd.concat([df_TRUST4, df_temp_aa], ignore_index=True, sort=False)
    df_TRUST4['tool'] = 'TRUST4'
```

```python
df_combined = pd.concat([df_MIXCR,df_IMREP,df_TRUST4], ignore_index=True, sort=False)
df_combined['th'] = df_combined['th']
df_combined
```

## sample01

```python
df_sample01 = df_combined.loc[df_combined['sample'] == 'sample01']

sample01_MIXCR = df_sample01.loc[df_sample01['tool'] == 'MIXCR']
sample01_IMREP = df_sample01.loc[df_sample01['tool'] == 'IMREP']
sample01_TRUST4 = df_sample01.loc[df_sample01['tool'] == 'TRUST4']

sns.set_style("white") 
sns.set_context("talk")  

fig, ax = plt.subplots(figsize=(5, 4))
sns.lineplot(sample01_MIXCR['th'], sample01_MIXCR['observed_portion_frequency'],label='MIXCR',color='#4878d0')
sns.lineplot(sample01_IMREP['th'], sample01_IMREP['observed_portion_frequency'],label='IMREP',color='#ee854a')
sns.lineplot(sample01_TRUST4['th'], sample01_TRUST4['observed_portion_frequency'],label='TRUST4',color='#6acc64')

ax.legend(['MIXCR','IMREP','TRUST4'])

plt.legend(bbox_to_anchor=(1,1), frameon=False, ncol=1, handletextpad=0.2, columnspacing=0.8, handlelength=1, prop={'size': 15})
locs, labels = plt.xticks()
ax.set_ylabel('Portion of repetoire captured by tool')
ax.set_xlabel('Clonotype frequency (log)')
ax.set_xscale('log')
plt.title('sample01',y=1.1,fontsize=16)
ax.set_ylim(0.925,1)

# setting label sizes after creation
ax.xaxis.label.set_size(16)
ax.yaxis.label.set_size(16)
ax.tick_params(axis='both', which='major', labelsize=13)
ax.tick_params(axis='both', which='minor', labelsize=13)
plt.xticks(rotation=45)
sns.despine()
plt.subplots_adjust(wspace=0.5)

plt.savefig("../figures/supplementary/figureS8/figureS8_a1.pdf",bbox_inches="tight")
plt.savefig("../figures/supplementary/figureS8/figureS8_a1.png",bbox_inches="tight")
plt.show(block=False)
```
