```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.ticker as mtick
```

```python
df = pd.read_csv("../summary_data/original/all_tools_TRB.csv")
df = df.loc[~(df['frequency_TCR'] == 0)]
df
```

```python
# Define clonotype groups based on frequency
def clonotype_group (row):
    if row['frequency_TCR'] > 0.1 and row['frequency_TCR'] <= 1:
        return 'Hyperexpanded'
    if row['frequency_TCR'] > 0.01 and row['frequency_TCR'] <= 0.1: 
        return 'Large'
    if row['frequency_TCR'] > 0.001 and row['frequency_TCR'] <= 0.01:
        return 'Medium'
    if row['frequency_TCR'] > 0.0001 and row['frequency_TCR'] <= 0.001:
        return 'Small'
    if row['frequency_TCR'] > 0 and row['frequency_TCR'] <= 0.0001:
        return 'Rare'

# Apply the clonotype_group function to the dataframe
df['clonotype_group'] = df.apply (lambda row: clonotype_group(row),axis=1)
```

```python
TCR_abundance= df.groupby(['Sample','tissue_type','repertoire_type','clonotype_group']).agg({'frequency_TCR':'sum'}).reset_index().rename(columns={'frequency_TCR':"clonotype_abundance"})
TCR_abundance.loc[:,'clonotype_abundance']=TCR_abundance['clonotype_abundance']*100
TCR_abundance
```

```python
mean = TCR_abundance.groupby(["repertoire_type",'clonotype_group'])["clonotype_abundance"].agg(["mean", "std"]) 
display(mean)
```
